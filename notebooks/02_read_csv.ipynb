## Opening CSV from URL

```python
import io
import requests

# Paste online link 
url = "https://raw.githubusercontent.com/MainakRepositor/Datasets-/master/zomato.csv"

# Setting headers prevents websites from blocking request
headers = {
    "User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"
}

# Download the data from the website
req = requests.get(url, headers=headers)

# Use io.StringIO to convert the text data into an in-memory file object
data = io.StringIO(req.text)

# Load the file object straight into a pandas DataFrame
# Read the file using standard latin-1 encoding to prevent character errors
df = pd.read_csv(data, encoding="latin-1")

# Preview  imported online data
df.head()
```

### squeeze Parameter

```python
""" To get the exact same behavior as squeeze=True (which forces a single-column DataFrame to become a Series), 
    append .squeeze() directly to the end.
"""

"""1. usecols=['enrollee_id', 'target'] --> loads a 2-column DataFrame.
   2. index_col='enrollee_id' --> moves the IDs to the row index, leaving exactly 1 column (target) in the main DataFrame body.
   3. squeeze() --> detects that single remaining data column and successfully collapses the whole object into a clean Pandas Series.
"""

df1 = pd.read_csv('../data/aug_train.csv',index_col='enrollee_id' ,usecols =['enrollee_id', 'target']).squeeze()
df1
```

```python
# Defining mapping dictionary for the teams
team_map = {
    "Royal Challengers Bangalore": "RCB",
    "Royal Challengers Bengaluru": "RCB",
    "Mumbai Indians": "MI",
    "Chennai Super Kings": "CSK",
    "Kolkata Knight Riders": "KKR",
    "Rajasthan Royals": "RR",
    "Delhi Capitals": "DC",
    "Delhi Daredevils": "DC",
    "Kings XI Punjab": "KXIP",
    "Punjab Kings": "PBKS",
    "Sunrisers Hyderabad": "SRH",
    "Deccan Chargers": "DCG",
    "Gujarat Titans": "GT",
    "Lucknow Super Giants": "LSG",
}

# Creating a converter function that replaces names or falls back to the original
def shorten_team(team_name):
    # .strip() removes accidental leading/trailing spaces
    return team_map.get(team_name.strip(), team_name)

import pandas as pd

# If the text matches (even with spaces), change it to 'IPL', otherwise keep original
ipl_converter = lambda text: "IPL" if text.strip().lower() == "indian premier league" else text


#  Read the file passing your function into the converters dictionary
df_ipl = pd.read_csv("../data/IPL.csv", low_memory=False, parse_dates=['date'], converters={"batting_team": shorten_team, "bowling_team":shorten_team, "event_name": ipl_converter})


# 4. Check the unique shortened names
df_ipl.head()
```
