```python
import numpy as np
import torch
```

## 0D Tensor(Scalar)

```python
# Numpy construction
np_scalar = np.array(56)

# PyTorch construction
pt_scalar = torch.tensor(56)

print(f"Numpy-Shape :{np_scalar.shape} | Dimensions : {np_scalar.ndim}")
print(f"Pytorch-Shape :{pt_scalar.shape} | Dimensions :{pt_scalar.dim()}")
```

## 1D Tensor(Vector)

```python
# Numpy construction (3-Dimensional Feature Vector)
np_vector = np.array([4.0, 5.6, 8.0])

# PyTorch construction
pt_vector = torch.tensor([4.0, 5.6, 8.0])

print(f"Numpy-Shape :{np_vector.shape} | Dimensions : {np_vector.ndim}")
print(f"Pytorch-Shape :{pt_vector.shape} | Dimensions :{pt_vector.dim()}")
```

## 2D Tensor(Matrix)


```python
# Matrix Data : 2 samples (rows) x 3 feature (column)
matrix = [[5.0, 6.0, 5.5],
          [6.0, 3.5, 2.1]]

# Numpy construction 
np_matrix = np.array(matrix)

# PyTorch construction
pt_matrix = torch.tensor(matrix)

print(f"Numpy-Shape :{np_matrix.shape} | Dimensions : {np_matrix.ndim}")
print(f"Pytorch-Shape :{pt_matrix.shape} | Dimensions :{pt_matrix.dim()}")
```

## 3D Tensor(Sequence/Time-series)

```python
# Simulating Text Data: Batch size of 2, 3 words per sentence, 4 embedding features per word
# Structural Shape Target: (2, 3, 4)


np_3d = np.random.rand(2, 3, 4)
pt_3d = torch.rand(2, 3, 4)


print(f"NumPy 3D Shape: {np_3d.shape} | Dimensions : {np_3d.ndim}")
print(f"PyTorch 3D Shape: {pt_3d.shape} | Dimension : {pt_3d.dim()}")
```

## 4D Tensor(Image Batches)

```python
# Simulating Computer Vision Data: Batch of 32 images, 64x64 resolution, 3 channels (RGB)
# Structural Shape Target: (Batch, Height, Width, Channels) for NumPy/TensorFlow


np_images = np.random.randint(0, 256, size=(32, 64, 64, 3), dtype=np.uint8)

# Structural Shape Target: (Batch, Channels, Height, Width) preferred convention for PyTorch
pt_images = torch.randint(0, 256, size=(32, 3, 64, 64), dtype=torch.uint8)

# Inspection
print(f"NumPy Image Batch Shape: {np_images.shape} | Dimensions : {np_images.ndim}")
print(f"PyTorch Image Batch Shape: {pt_images.shape} | Dimensions : {pt_images.dim()}")
```

## 5D Tensor(Video Batches)

```python
# Simulating Video Data: Batch of 4 clips, each has 30 frames, 128x128 resolution, 3 color channels
# Structural Shape Target: (Batch, Frames, Channels, Height, Width)

np_videos = np.random.randint(0, 256, size=(4, 30, 3, 128, 128), dtype=np.uint8)
pt_videos = torch.rand(4, 30, 3, 128, 128)



print(f"Numpy Video Batch Shape: {np_videos.shape}")
print(f"Total scalar items inside video stream array: ","\n")

print(f"PyTorch Video Batch Shape: {pt_videos.shape}")
print(f"Total scalar items inside video stream tensor: {pt_videos.numel():,}")
```
