## Opening json from URL

````python
import io
import pandas as pd  # Added missing pandas import
import requests

# Fixed: Removed the extra space at the very end of the URL string
url = "https://api.exchangerate-api.com/v4/latest/INR"

# header prevents website from blocking request
headers = {
    "User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"
}

# download the data from website
req_data = requests.get(url, headers=headers)

# Use .json() to parse the response text directly into Python dictionaries/lists
json_data = req_data.json()

# Parse the inner 'rates' dictionary so your DataFrame looks like a clean table
df = pd.DataFrame(list(json_data["rates"].items()), columns=["Currency", "Rate"])

# Preview imported online data
df.info()
print("---"*20)
print(df.head(5))
````

## Setting Up a Local MySQL Database on Mac (DBngin + TablePlus)

For using the `world.sql` lets host this on local database


### Step 1:- Install the tools

- **[DBngin](https://dbngin.com/)** — a free, lightweight database manager for Mac. Runs MySQL, PostgreSQL, or Redis instances locally without messing with Homebrew services or config files.
- **[TablePlus](https://tableplus.com/)** — a GUI client for browsing, querying, and importing data into databases. Free tier is enough for this workflow.

Download both as `.dmg` files from their official sites and drag them into `Applications`.

### Step 2:- Create a database instance in DBngin

1. Open DBngin → click the **+** button → select **MySQL**
2. Choose a version (8.0 recommended) and keep the default port (`3306`)
3. Click **Create**, then hit **Start** — the instance turns green when running

### Step 3:- Connect TablePlus to the local server

1. Open TablePlus → right-click the empty pane → **Create a new connection** → **MySQL**
2. Enter:
   - **Host:** `127.0.0.1`
   - **Port:** `3306`
   - **User:** `root`
   - **Password:** *(leave blank — DBngin's default root has no password)*
3. Click **Test**, then **Connect**

### Step 4:- Import the sample dataset

1. In TablePlus, right-click the connection → **New Database** → name it `world`
2. Select the `world` database → **File → Import → From SQL Dump...**
3. Choose the downloaded `world.sql` file
4. Run the import — this creates and populates the `city`, `country`, and `countrylanguage` tables

Verify with a quick query in TablePlus:
```sql
SELECT * FROM city LIMIT 5;
```

````python
from sqlalchemy import create_engine
import pandas as pd

engine = create_engine("mysql+pymysql://root:@127.0.0.1:3306/world")
df = pd.read_sql("SELECT * FROM city LIMIT 10", engine)
df.info()
df.head(5)

````

## ⚖️ SQLAlchemy vs mysql.connector — Which to Use Daily

Both connect to MySQL fine, but they serve different purposes. Here's the breakdown:

| | **SQLAlchemy + PyMySQL** | **mysql.connector** |
|---|---|---|
| **Type** | ORM/engine layer | Raw official MySQL driver |
| **pandas support** | Native — `pd.read_sql()` works cleanly, no warnings | Works, but throws `UserWarning` |
| **Connection object** | `engine` — reusable, supports pooling | `conn` — single raw connection |
| **Querying** | Direct via `pd.read_sql(query, engine)` | Needs a `cursor` for raw queries |
| **Best for** | Data pipelines, repeated queries, pandas-heavy workflows | Quick one-off scripts, fine-grained cursor control, admin-style operations |
| **Industry standard for ML/data work** | ✅ Yes | ❌ Less common |

### ✅ Daily driver: `sqlalchemy + pymysql`

For this pipeline — and for ML/data work in general — **SQLAlchemy is the default**. Reasons:
- Every `pd.read_sql()` call is clean, no warning noise
- The `engine` object is reusable across the notebook without reconnecting
- It's what most tutorials, libraries, and production codebases expect
- Swapping databases later (e.g. MySQL → PostgreSQL) only means changing the connection string, not the code logic

### When `mysql.connector` still earns its place

Keep it in your back pocket for:
- Debugging a connection issue at the raw driver level
- Running a quick admin query without spinning up an engine
- Situations where you need fine control over cursors (e.g. multi-statement execution, stored procedures)

**Rule of thumb:** if it's going into a DataFrame, use SQLAlchemy. If it's a one-off sanity check, either works — but default to SQLAlchemy for consistency.
