### 10. Survival rate by category 
— Pearson correlation isn't the right tool for categorical columns like Sex, Pclass, Embarked. Use groupby instead.

```python
# Correlation survival with respect to sex
# Female survival is larger as priority was "women and children first" 
df1.groupby('Sex')['Survived'].mean()
```

```python
# Correlation of Pclass with respect to Survival 
df1.groupby('Pclass')['Survived'].mean()
```

```python
# Correlation of Embarked with respect to Survival
df1.groupby('Embarked')['Survived'].mean()
```

Sex turns out to be the single strongest signal in this dataset (74% vs 19% survival) — it got missed in the original notebook because it was dropped *before* checking correlation.

### 11. Fare — checking skew/outliers, not just mean/std

```python
df1['Fare'].skew()
```

A skew of ~4.8 is heavily right-skewed (0 = symmetric). Combined with std (49.7) being larger than the mean (32.2) from `describe()`, this confirms outliers — there's a fare as high as 512 against a median of just 14.45. Worth a log-transform or capping before modeling.

### 12. Cabin is 77% missing — instead of dropping it outright, turn it into a signal

```python
df1['Has_Cabin'] = df1['Cabin'].notnull().astype(int)
df1.groupby('Has_Cabin')['Survived'].mean()
```

Having a recorded cabin is itself predictive (67% vs 30% survival) — likely a proxy for class/fare. This turns a mostly-missing column into a usable binary feature instead of throwing it away.
